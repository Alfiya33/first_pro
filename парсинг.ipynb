Парсинг данных: достаю из сайта данные о профессии архитектор:

```python
import requests
URL='https://api.hh.ru/vacancies'
par1={'text': "Архитектор",
      'area': 1,
      'page': 0,

}
data1=requests.get(URL,par1).json()
data1
```

Сосчитала количество вакансии всего:

```python
amount_1=data1.get('found', 0)
amount_1
```

Создала  структуру данных, которая  хранить в себе имя компании и ее координаты (широта и долгота):

```python
import pandas as pd 
import requests
company1_data=[]

for vac in data1.get('items', []): # данная строка перебирает вакансия из data1 и, если items пустой, выйдет просто нуловое значение 
    name_company=vac.get('employer', {}).get('name', 'не указано')
    if 'address' in vac and vac['address']:
        lat=vac['address'].get('lat','не указано')
        lon=vac['address'].get('lon', 'не указано')
    else:
        lat='не указано'
        lon='не указано'
    company1_data.append(
        { 'Имя компании':name_company,
         'Широта': lat,
         'Долгота': lon
        }
    )


df1=pd.DataFrame(company1_data)
df1
```
